```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample job data with Job Portal and Company
data = {
    'Job Portal': ['Indeed', 'LinkedIn', 'Naukri', 'Indeed', 'LinkedIn', 'Monster', 'Naukri', 'Indeed'],
    'Company': ['TCS', 'Infosys', 'Wipro', 'Accenture', 'Capgemini', 'Tech Mahindra', 'HCL', 'IBM']
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Display the data
df

# Count the number of companies per Job Portal
portal_counts = df['Job Portal'].value_counts()
portal_counts
```

```python
# Plotting the bar chart
plt.figure(figsize=(8, 5))
portal_counts.plot(kind='bar', color='skyblue', edgecolor='black')

plt.title('Job Portal vs Company Count')
plt.xlabel('Job Portal')
plt.ylabel('Number of Companies')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(rotation=0)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import folium
import datetime
from IPython.display import display, HTML

data = [
    {
        'Job Portal': 'Indeed',
        'Company': 'TCS',
        'Qualification': 'B.Tech',
        'Work Type': 'Full time',
        'Country': 'Nigeria',
        'Job Title': 'Data Scientist',
        'Preference': 'Male',
        'Company Size': 100000,
        'Contact Person': 'Amit',
        'Latitude': 9.0820,
        'Longitude': 8.6753
    },
    {
        'Job Portal': 'LinkedIn',
        'Company': 'Infosys',
        'Qualification': 'MBA',
        'Work Type': 'Full time',
        'Country': 'Kenya',
        'Job Title': 'Developer',
        'Preference': 'Female',
        'Company Size': 90000,
        'Contact Person': 'Rahul',
        'Latitude': -1.2921,
        'Longitude': 36.8219
    },
    {
        'Job Portal': 'Indeed',
        'Company': 'IBM',
        'Qualification': 'PhD',
        'Work Type': 'Full time',
        'Country': 'Egypt',
        'Job Title': 'Developer',
        'Preference': 'Male',
        'Company Size': 85000,
        'Contact Person': 'Anjali',
        'Latitude': 30.0444,
        'Longitude': 31.2357
    }
]

df = pd.DataFrame(data)
df

```

```python

# Get IST current time
now = datetime.datetime.utcnow() + datetime.timedelta(hours=5, minutes=30)
hour = now.hour

# Define valid African countries
african_countries = ['Nigeria', 'Kenya', 'Egypt', 'Ghana', 'South Africa', 'Morocco']

if 15 <= hour <= 18:
    # Apply filters
    filtered = df[
        (df['Qualification'].isin(['B.Tech', 'M.Tech', 'PhD'])) &
        (df['Work Type'] == 'Full time') &
        (df['Country'].isin(african_countries)) &
        (df['Job Title'].str.startswith('D')) &
        (df['Preference'] == 'Male') &
        (df['Company Size'] > 80000) &
        (df['Contact Person'].str.startswith('A')) &
        (df['Job Portal'] == 'Indeed')
    ]

    # Show result or message
    if filtered.empty:
        print("No matching data found with all filters.")
    else:
        display(filtered)

        # Bar chart for companies
        company_counts = filtered['Company'].value_counts()
        plt.figure(figsize=(8, 5))
        company_counts.plot(kind='bar', color='green', edgecolor='black')
        plt.title('Filtered Companies - Task 2')
        plt.xlabel('Company')
        plt.ylabel('Count')
        plt.xticks(rotation=0)
        plt.grid(axis='y', linestyle='--')
        plt.show()

        # Show map links
        for index, row in filtered.iterrows():
            lat = row['Latitude']
            lon = row['Longitude']
            link = f"https://www.google.com/maps?q={lat},{lon}"
            display(HTML(f"<b>{row['Company']}</b>: <a href='{link}' target='_blank'>Open on Google Maps</a>"))

else:
    print("Task 2 chart is only visible between 3 PM and 6 PM IST.")
```
